# Random Forest Regressor

```python
from sklearn.model_selection import train_test_split
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, test_size=0.20) #80% data will be for training and 20% for testing.
```

```python
from sklearn.model_selection import GridSearchCV #GridSearchCV is for parameter tuning
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
n_estimators=[103,104,105,106,107,108] #number of decision trees in the forest, default = 100
criterion=['mse'] 
max_depth=[11,12,13,14,15,16,17]
parameters={'n_estimators': n_estimators,'criterion':criterion,'max_depth':max_depth} 
```

```python
RFR_reg = GridSearchCV(regressor,parameters)
RFR_reg.fit(X_train2,y_train2)
```

```python
RFR_reg.best_params_
```

```python
ypred2 = RFR_reg.predict(X_test2)
```

```python
from sklearn.metrics import mean_squared_error #calculating MSE
MSE=mean_squared_error(y_test2,ypred2)
print("MSE:",MSE)
```

```python
np.sqrt(mean_squared_error(y_test2,ypred2))
```

```python
result2=RFR_reg.score(X_test2,ypred2)
result2="{:.2f} %".format(result2.mean()*100)
result2
```
